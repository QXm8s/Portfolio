## Описание проекта
Оператор связи «Ниединогоразрыва.ком» предоставляет 2 вида основных услуг: 
- стационарную телефонную связь (возможно одновременное подключение телефонного аппарата к нескольким линиям), 
- интернет (DSL или Fiber optic). 

А также дополнительные услуги:
- антивирус и межсетевой экран, 
- выделенная линия технической поддержки, 
- облачное хранилище для резервного копирования данных, 
- стриминговое телевидение и каталог фильмов.

Клиенты могут оплачивать услуги ежемесячно или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного счета на оплату.

Компания хочет прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Одновременно должны снизиться убытки от неэффективных акций по удержанию клиентов. 

### Постановка задачи
Требуется построить модель, которая предскажет уход клиента с метрикой качества ROC-AUC не менее 0,85. Для интерпретации результатов использовать ROC-кривую, Accuracy, Confusion Matrix и Feature Importance. 

### Описание данных 
Команда оператора собрала обезличенные данные о клиентах, информацию об их тарифах и договорах.
Данные состоят из файлов, полученных из разных источников:
 - contract_new.csv (информация о договоре):
     - BeginDate – дата начала пользования услугами, 
     - EndDate – дата окончания пользования услугами,  
     - Type – тип договора: ежемесячный, годовой и т.д., 
     - PaperlessBilling – выставления счёта по электронной почте, 
     - PaymentMethod – способ оплаты, 
     - MonthlyCharges – ежемесячные траты на услуги, 
     - TotalCharges – всего потрачено на услуги.
 - personal_new.csv (данные клиента): 
     - gender – пол, 
     - SeniorCitizen – пенсионный статус по возрасту,  
     - Partner – наличие супруга/супруги, 
     - Dependents – наличие иждивенцев.      
 - internet_new.csv (информация об интернет-услугах): 
     - InternetService – наличие услуг Интернет, 
     - OnlineSecurity – межсетевой экран, 
     - OnlineBackup – облачное хранилище для резервного копирования данных, 
     - DeviceProtection – антивирус, 
     - TechSupport - выделенная линия технической поддержки, 
     - StreamingTV – онлайн-ТВ, 
     - StreamingMovies – онлайн-кинотеатр.
 - phone_new.csv — информация об услугах телефонии: 
     - MultipleLines – возможность подключения телефонного аппарата к нескольким линиям одновременно

**Важно:** 
 - во всех файлах столбец `customerID` содержит код клиента, 
 - информация о договорах актуальна на 1 февраля 2020.

Сформулируем приемрный ход решения:

Признак `EndDate` содержит, как даты, так и значение `No`, которое указывает на действующие (не расторгнутые) договоры. 

Признак `TotalCharges`, явно содержащий численные данные, определился, как `object`, что указывает на наличие значений, неприводимых к численному типу. Проверим это, отсортировав значения признака.

 - для новых абонентов установим `TotalCharges = 0`, 
 - для действующих абонентов заменим `EndDate` с `No` на дату выгрузки (2020-02-01),

```python
#преобразуем типы
types = {'BeginDate' : 'datetime64', 
         'EndDate' : 'datetime64',
         'TotalCharges' : 'float64'}
df_contract = df_contract.astype(types)
df_contract.sample(5)
```

Предварительно опишем функции для построения диаграмм. 

```python
def wedge_label(pct, values):
    """
    Возвращает строку-подпись для сектора круговой 
    диаграммы с долей в процетах и абсолютным значением. 
    Например: 46% (3794)
    АРГУМЕНТЫ: 
    -- pct - доля в процентах для текущего сектора, 
    -- values - список абсолютных значений для всех секторов 
    """
    absolute = int(np.round(pct/100.*np.sum(values)))
    return '{0}%\n({1})'.format(round(pct, 1), absolute)
```

```python
def plot_pie(data, feature): 
    """
    Строит круговые диаграммы категориального признака
    feature для сегментов: 
     - 'Все клиенты' (discontinued = None), 
     - 'Бывшие клиенты' (discontinued = Yes), 
     - 'Действующие клиенты' (discontinued = No).     
    АРГУМЕНТЫ: 
    -- data - DataFrame с данными, 
    -- feature - признак, для которого строятся диаграммы.
    """    
    segment_feature = 'discontinued'
    segment_values = [
        None, 
        'Yes', 
        'No'
    ]
    segment_aliases = [
        'Все клиенты', 
        'Бывшие клиенты', 
        'Действующие клиенты'
    ]
    ncols = len(segment_values)     
    fig, ax = plt.subplots(1, ncols, figsize=(20, 6))
    for col in range(0, ncols): 
        segment_value = segment_values[col] 
        segment_alias = segment_aliases[col]                
        
        x = None
        if segment_value is None:
            x = data[feature].value_counts().sort_index()            
        else: 
            exp = data[segment_feature] == segment_value
            x = data[exp][feature].value_counts().sort_index()
        
        values = x.values.tolist() 
        labels = x.index.tolist() 
        ax[col].pie(
            x=values, 
            labels=labels, 
            autopct=lambda pct : wedge_label(pct, values), 
            explode=[0.025] * len(labels)
        )         
        ax[col].set_title(segment_alias)
        
    fig.suptitle('Распределение признака \'{0}\''.format(feature))
    fig.show()        
```

```python
def plot_hist(data, feature, feature_alias, exclude=None): 
        """
        Строит гистограммы количественного признака
        feature для сегментов:
         - 'Все клиенты' (discontinued = None), 
         - 'Бывшие клиенты' (discontinued = Yes), 
         - 'Действующие клиенты' (discontinued = No). 
         АРГУМЕНТЫ: 
         -- data - DataFrame с данными, 
         -- feature - признак, для которого строятся гистограммы, 
         -- feature_alias - подпись для оси X, 
         -- exclude - дополнительный фильтр для признака feature.
         """    
        segment_feature = 'discontinued' 
        segment_values = [
            None, 
            'Yes', 
            'No'
        ] 
        segment_aliases = [
            'Все клиенты', 
            'Бывшие клиенты', 
            'Действующие клиенты'
        ]
        nrows = len(segment_values)     
        fig, ax = plt.subplots(nrows, 1, figsize=(16, 13))
        for row in range(0, nrows): 
            segment_value = segment_values[row] 
            segment_alias = segment_aliases[row] 
            
            x = None
            exp = None
            if segment_value is None:
                if exclude is None: 
                    x = data[feature]
                else: 
                    exp = data[feature] != exclude
                    x = data[exp][feature]                
            else:                 
                if exclude is None:
                    exp = data[segment_feature] == segment_value
                else: 
                    exp = (data[feature] != exclude) \
                    & (data[segment_feature] == segment_value)
                x = data[exp][feature]
                                
            ax[row].hist(x=x, bins=150)
            ax[row].set_xlabel(feature_alias) 
            ax[row].set_ylabel('Кол-во записей, шт')         
            ax[row].grid(True) 
            ax[row].set_title(segment_alias) 
            
        fig.suptitle('Распределение признака \'{0}\''.format(feature))
        fig.tight_layout()
        fig.show()
```

На основании корреляционного анализа признаков ([см. 4.4](#%D0%9A%D0%BE%D1%80%D1%80%D0%B5%D0%BB%D1%8F%D1%86%D0%B8%D1%8F-%D0%BF%D1%80%D0%B8%D0%B7%D0%BD%D0%B0%D0%BA%D0%BE%D0%B2)) отберем для обучения и валидации моделей наиболее значимые, а именно: 
 - длительность обслуживания (`durability`),
 - ежемесячные траты (`monthly_charges`)
 - семейный статус (`partner`), 
 - способ оплаты (`payment_method`),
 - телефония (`multiple_lines`). 
 
Признак общих трат (`total_charges`) исключим во избежание негативных эффектов мультиколлинеарности. 

Разделим отобранные признаки на группы: `fts_to_encode` - для категориальных и `fts_to_scale` - для количественных.

```python
fts_to_scale = [    
    'monthly_charges',     
    'durability'
]
fts_to_encode = [        
    'partner', 
    'payment_method',     
    'multiple_lines'
]    
```

### Формирование выборок
Перед нами - задача бинарной классификации в условиях дисбаланса классов. Следовательно, соотношение классов в основном датасете должно воспроизводиться в выборках. Для этого сформируем стратифицированные обучающую и тестовую выборки.

```python
X = df[fts_to_encode + fts_to_scale]
y = df[TARGET].replace({'Yes' : 1, 'No' : 0})

X_train, X_test, y_train, y_test = \
train_test_split(X, y, test_size=0.25, stratify=y, random_state=SEED)
```

### Обучение и валиадция моделей
На обучающей выборке, с кросс-валидацией и метрикой качества AUC-ROC: 
- закодируем и отмасштабируем признаки, 
- найдем оптимальные гиперпараметры для моделей: 
     - LogisticRegression, 
     - RandomForestClassifier,
     - CatBoostClassifier;
- дисбаланс классов дополнительно компенсируем параметром `class_weight = 'balanced'`,
- определим лучшую модель.

Воспользуемся преимуществом CatBoostClassifier: обучим и валидируем модель без предварительного кодирования и масштабирования, передав параметр `cat_features`. 

```python
def make_col_transformer(to_encode, to_scale):
    '''Возвращает ColumnTransformer для 
    кодирования и масштабирования признаков. 
    Примечания:
     - признаки, не подлежащие кодированию или 
    масштабированию остаются в выборке (remainder='passthrough').
    АРГУМЕНТЫ:
     -- to_encode- признаки, подлежащие кодированию, 
     -- to_scale - признаки, подлежащие масштабированию.
    '''        
    encoder = OneHotEncoder(drop='first', handle_unknown='ignore')
    scaler = MinMaxScaler()    
    transformer = ColumnTransformer(
        [
            ('encoder', encoder, to_encode), 
            ('scaler', scaler, to_scale)
        ], 
        remainder='passthrough', 
        verbose_feature_names_out=False
        )        
    return transformer
```

```python
#инициализируем pipeline и его параметры
pipeline = Pipeline(
    [
        ('transformer', None),
        ('model', None)
    ]
)
params = [
    #кодирование и масштабировние + LogisticRegression 
    {
        'transformer' : [make_col_transformer(fts_to_encode, fts_to_scale)], 
        'model': [LogisticRegression(
            class_weight='balanced', 
            random_state=SEED
        )], 
        'model__C' : [1, 10, 15],
        'model__max_iter' : [100, 200],
    }, 
    #кодирование и масштабировние + RandomForestClassifier
    {
        'transformer' : [make_col_transformer(fts_to_encode, fts_to_scale)], 
        'model': [RandomForestClassifier(
            class_weight='balanced', 
            random_state=SEED)],
        'model__max_depth' : [7, 14, 20],                
        'model__n_estimators' : [100, 200], 
    }, 
    #CatBoostClassifier без кодирования и масштабировния 
    {
        'transformer' : [None],          
        'model': [CatBoostClassifier(
            cat_features=fts_to_encode, 
            auto_class_weights = 'Balanced', 
            random_state=SEED, 
            verbose=False
        )],        
        'model__learning_rate' : [0.25, 0.5, 0.75],                
        'model__iterations' : [500, 1000], 
    } 
]
#инициализируем механизм поиска
search = GridSearchCV(
    pipeline, 
    param_grid=params,
    scoring='roc_auc', 
    verbose=0
)
```

```python
%%time
search.fit(X_train, y_train)
```

Посмотрим на лучшие результаты для каждого класса моделей.

```python
#считываем результаты поиска
results = pd.DataFrame(search.cv_results_)
results['est_class_name'] = \
results['param_model'].apply(lambda x : x.__class__.__name__)
#готовим сводную таблицу лучших моделей каждого класса
results_score = results.groupby('est_class_name') \
                    .agg({'mean_test_score' : 'max'}) \
                    .sort_values(by='mean_test_score', ascending=False)
results_score.index.name = None 
results_score.columns = ['ROC-AUC (validation)']
results_score
```

Применительно к обучающей выборке: 
 - лучшая модель - CatBoostClassifier (ROC-AUC = 0.90),
 - на втором месте - RandomForestClassifier (ROC-AUC = 0.80), 
 - замыкает тройку - LogisticRegression (ROC-AUC = 0.71), 
 - RandomForestClassifier и LogisticRegression не преодолели порогового значения метрики (ROC-AUC = 0.85). 

Определим ROC-AUC для лучшей модели на тестовой выборке.

```python
y_pred_ones_test = search.predict_proba(X_test)[:,1]
roc_auc_test = roc_auc_score(y_test, y_pred_ones_test)
all_params = search.best_estimator_['model'].get_all_params()
params = {
    'learning_rate' : all_params['learning_rate'], 
    'iterations' : all_params['iterations']
}
print(
    'Тестовая выборка:',
    '\n - модель', search.best_estimator_['model'].__class__.__name__, 
    '\n - ROC-AUC', round(roc_auc_test, 6), 
    '\n - гиперпараметры:', params, 
    '\n - признаки:', X_test.columns.tolist()
) 
```

Т.о., CatBoostClassifier, обученный на признаках `partner, payment_method, multiple_lines, monthly_charges, durability` с параметрами `learning_rate = 0.25, iterations = 1000` демонстрирует метрику качества, удовлетворяющую условиям задачи (ROC-AUC >= 0.85): 
- ROC-AUC = 0.90 на валидационной выборке, 
- ROC-AUC = 0.91 на тестовой выборке.
